```python
import numpy as np
import pandas as pd

def _num_report(x: pd.Series):
    x_nonnull = x.dropna()
    n = len(x)
    n_nonnull = len(x_nonnull)
    miss_rate = (n - n_nonnull) / n if n else np.nan

    # 基本统计
    skew = x_nonnull.skew() if n_nonnull else np.nan
    kurt = x_nonnull.kurt() if n_nonnull else np.nan

    # IQR 异常值比例
    outlier_rate = np.nan
    if n_nonnull >= 5:
        q1, q3 = np.percentile(x_nonnull, [25, 75])
        iqr = q3 - q1
        lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
        outlier_rate = ((x_nonnull < lower) | (x_nonnull > upper)).mean()

    # 长尾/偏态判断
    tail_ratio = np.nan
    if n_nonnull >= 5:
        p95 = np.percentile(x_nonnull, 95)
        p50 = np.percentile(x_nonnull, 50)
        tail_ratio = (p95 / (p50 + 1e-9)) if (p50 + 1e-9) != 0 else np.inf

    hints = []
    # 偏态判定阈值：|skew| > 0.75 常见经验阈
    if pd.notna(skew) and abs(skew) > 0.75:
        if skew > 0:
            hints.append("右偏明显（长右尾）")
        else:
            hints.append("左偏明显（长左尾）")
    if pd.notna(tail_ratio) and tail_ratio > 1.8:
        hints.append("重尾/长尾（P95 远高于中位数）")
    if pd.notna(outlier_rate) and outlier_rate > 0.05:
        hints.append(f"异常值较多（约 {outlier_rate:.1%}）")

    # 建议
    recs = []
    if miss_rate > 0:
        recs.append("缺失值：数值型可用中位数、0（表示无该特征）或邻域/模型插补")
    if abs(skew) > 0.75:
        # 针对非负且右偏推荐 log1p / Box-Cox；其他推荐 Yeo-Johnson
        if (x_nonnull.min() >= 0) and (skew > 0):
            recs.append("分布：右偏→尝试 log1p / Box-Cox；也可试 Yeo-Johnson")
        else:
            recs.append("分布：偏态→尝试 Yeo-Johnson 变换")
    if pd.notna(outlier_rate) and outlier_rate > 0.05:
        recs.append("异常值：IQR 截尾/Winsorize 或使用 RobustScaler")
    if pd.notna(tail_ratio) and tail_ratio > 1.8:
        recs.append("长尾：对数/分桶（如等频分箱）减少重尾影响")
    if not recs:
        recs.append("分布整体可接受：可直接标准化/归一化或交给树模型")

    return {
        "type": "numeric",
        "non_null": int(n_nonnull),
        "missing_rate": round(float(miss_rate), 6) if pd.notna(miss_rate) else np.nan,
        "skew": round(float(skew), 6) if pd.notna(skew) else np.nan,
        "kurtosis": round(float(kurt), 6) if pd.notna(kurt) else np.nan,
        "outlier_rate_iqr": round(float(outlier_rate), 6) if pd.notna(outlier_rate) else np.nan,
        "tail_ratio_p95_to_median": round(float(tail_ratio), 6) if pd.notna(tail_ratio) else np.nan,
        "flags": hints,
        "recommendations": recs
    }

def _cat_report(x: pd.Series, rare_thresh=0.02):
    x_nonnull = x.dropna().astype(str)
    n = len(x)
    n_nonnull = len(x_nonnull)
    miss_rate = (n - n_nonnull) / n if n else np.nan

    vc = x_nonnull.value_counts(dropna=False)
    k = len(vc)

    # 最大类别占比 & 不平衡度
    max_cat = vc.index[0] if k > 0 else None
    max_prop = (vc.iloc[0] / n_nonnull) if n_nonnull else np.nan

    # 稀有类别集合
    rare_cats = vc[vc / n_nonnull < rare_thresh].index.tolist() if n_nonnull else []

    # 信息度量（香农熵/Gini）
    entropy = np.nan
    gini = np.nan
    if n_nonnull:
        p = (vc / n_nonnull).values
        entropy = -np.sum(p * np.log2(p + 1e-12))
        gini = 1 - np.sum(p ** 2)

    hints = []
    if miss_rate > 0:
        hints.append("存在缺失")
    if pd.notna(max_prop) and max_prop > 0.85:
        hints.append(f"极度不平衡（最大类别占比 {max_prop:.1%}）")
    if len(rare_cats) > 0:
        hints.append(f"稀有类别较多（<{rare_thresh:.0%} 占比）")

    recs = []
    if miss_rate > 0:
        recs.append('缺失值：用 "None"/"Missing" 或众数填补（若缺失代表无该属性，可保留为 "None"）')
    if len(rare_cats) > 0:
        recs.append('稀有类别：合并到 "Other" 或做频数编码（Frequency Encoding）')
    if pd.notna(max_prop) and max_prop > 0.85:
        recs.append("类别极不平衡：考虑目标编码/频数编码，或与其它特征交互；分类任务可用 class_weight/采样")
    # 默认编码建议
    recs.append("编码：树模型→可保持标签或做目标/频数编码；线性/距离模型→建议 One-Hot（合并稀有后）")

    return {
        "type": "categorical",
        "non_null": int(n_nonnull),
        "missing_rate": round(float(miss_rate), 6) if pd.notna(miss_rate) else np.nan,
        "n_categories": int(k),
        "top_category": str(max_cat) if max_cat is not None else None,
        "top_category_share": round(float(max_prop), 6) if pd.notna(max_prop) else np.nan,
        "n_rare_categories": int(len(rare_cats)),
        "rare_threshold": rare_thresh,
        "entropy_bits": round(float(entropy), 6) if pd.notna(entropy) else np.nan,
        "gini": round(float(gini), 6) if pd.notna(gini) else np.nan,
        "flags": hints,
        "recommendations": recs,
        "rare_categories_list_example": rare_cats[:10]
    }

def analyze_column(df: pd.DataFrame, col: str, rare_thresh: float = 0.02):
    """
    自动分析某列的分布并给出建议。
    rare_thresh: 对分类特征，低于该占比的类别视为稀有（默认 2%）
    """
    if col not in df.columns:
        raise ValueError(f"列 {col} 不存在")

    x = df[col]
    base = {
        "column": col,
        "dtype": str(x.dtype),
        "n_samples": int(len(x))
    }

    # 类型判定：object/类别→分类；其余尝试数值
    if pd.api.types.is_numeric_dtype(x):
        rep = _num_report(x)
    else:
        rep = _cat_report(x, rare_thresh=rare_thresh)

    rep.update(base)
    return rep

def pretty_print_report(report: dict):
    print(f"列名: {report.get('column')}  |  类型推断: {report.get('type')}  |  原始dtype: {report.get('dtype')}")
    print(f"样本数: {report.get('n_samples')}  |  非空数: {report.get('non_null')}  |  缺失率: {report.get('missing_rate'):.2%}")
    print("-"*60)
    if report.get("type") == "numeric":
        print(f"偏度: {report.get('skew')}, 峰度: {report.get('kurtosis')}")
        print(f"IQR异常值比例: {report.get('outlier_rate_iqr'):.2%}  |  P95/中位数: {report.get('tail_ratio_p95_to_median'):.3f}")
    else:
        print(f"类别数: {report.get('n_categories')}  |  最大类: {report.get('top_category')} ({report.get('top_category_share'):.2%})")
        print(f"稀有类别数(<{report.get('rare_threshold'):.0%}): {report.get('n_rare_categories')}")
        print(f"熵(比特): {report.get('entropy_bits')}  |  Gini: {report.get('gini')}")
        if report.get("n_rare_categories", 0) > 0:
            print(f"稀有类别示例: {report.get('rare_categories_list_example')}")
    if report.get("flags"):
        print("\n⚠️ 发现问题/提示：")
        for f in report["flags"]:
            print(" -", f)
    print("\n✅ 建议：")
    for r in report["recommendations"]:
        print(" -", r)

# —— 使用示例 ——
# 数值列：
# rep = analyze_column(df, "GrLivArea")
# pretty_print_report(rep)

# 分类列：
# rep = analyze_column(df, "Neighborhood", rare_thresh=0.02)
# pretty_print_report(rep)
```

数值型 vs. SalePrice：

散点图（如 GrLivArea vs SalePrice）。
